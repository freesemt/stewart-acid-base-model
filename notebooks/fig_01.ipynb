```
import numpy                as np
import matplotlib.pyplot    as plt
from matplotlib.ticker import FormatStrFormatter

Kw  = 4.4e-14
MEGA = 1e6

fig = plt.figure( figsize=(12, 5) )
fig.suptitle( "Strong Ion Solutions" )

ax1  = fig.add_subplot(121)
ax2  = fig.add_subplot(122)

for k, ax in enumerate([ax1, ax2]):
    lim = np.power( 10.0, -6+k ) 
    x   = np.linspace(-lim, lim, 100 )
    ax.set_title( "[$H^+$] and [$OH^⁻$] versus [SID]" )
    ax.set_xlabel( '[SID] (μequiv./L)' )
    ax.set_ylabel( '[$H^+$] or [$OH^⁻$] (μequiv./L)' )
    ax.xaxis.set_major_formatter(FormatStrFormatter('%.3g'))
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.3g'))
    ax.plot( MEGA*x, MEGA*( np.sqrt( Kw + (x/2)**2 ) - x/2 ), label='[$H^+$]' )
    ax.plot( MEGA*x, MEGA*( np.sqrt( Kw + (x/2)**2 ) + x/2 ), label='[$OH^⁻$]' )
    kw_ = MEGA*np.sqrt( Kw )
    xmin, xmax = ax.get_xlim()
    x_ = (xmax - xmin) * 0.05
    ax.annotate("[$H^+$]=[$OH^⁻$]=$\sqrt{K'_w}$", xy=(0, kw_), xytext=(x_, kw_), arrowprops=dict( arrowstyle="->"))
    ax.legend()

fig.tight_layout()
fig.subplots_adjust( top=0.88 )

plt.show()
```
