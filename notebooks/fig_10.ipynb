```
import numpy                as np
import matplotlib.pyplot    as plt
from matplotlib.ticker import FormatStrFormatter
import sys
sys.path.append('local_modules/OurUtils')
from StewartModel import Fh4
MILI = 1e-3

Atot = [ 10, 15, 20, 25 ]
SID = 42
pCO2 = np.linspace(10, 90, 100)

fig = plt.figure( figsize=(6, 5) )
fig.suptitle( "Standard Plasma" )

ax  = fig.add_subplot(111)
ax.set_title( "[$H^+$] versus $pCO_2$" )
ax.set_xlabel( "$pCO_2$ (mmHg)" )
ax.set_ylabel( "pH" )

for atot in Atot:
    H   = Fh4(SID*MILI, atot*MILI, pCO2 )
    pH  = -np.log10(H)
    ax.plot(pCO2, pH, label='[Atot] = %d mequiv./L' % atot)

pCO2_ = 41
nh  = Fh4(SID*MILI, 0.019, pCO2_ )
pH  = -np.log10(nh)
ax.plot( pCO2_, pH, '+', color='red', label='normal operating point' )

ax.legend()
fig.tight_layout()
fig.subplots_adjust( top=0.88 )

plt.show()
```
