```
import numpy                as np
import matplotlib.pyplot    as plt
from matplotlib.ticker import FormatStrFormatter
import sys
sys.path.append('local_modules/OurUtils')
from StewartModel import Fh4, Kc
MILI = 1e-3
KILO = 1e3

Atot = 0.019
SID = [ 30, 35, 40, 45, 50 ]
pCO2 = np.linspace(10, 90, 100)

fig = plt.figure( figsize=(6, 5) )
fig.suptitle( "Standard Plasma" )

ax  = fig.add_subplot(111)
ax.set_title( "[$HCO_3^⁻$] versus $pCO_2$" )
ax.set_xlabel( "$pCO_2$ (mmHg)" )
ax.set_ylabel( "[$HCO_3^⁻$] (mequiv./L)" )
ax.set_ylim( 0, 55 )

for sid in SID:
    H   = Fh4(sid*MILI, Atot, pCO2 )
    HCO3 = Kc*pCO2/H
    ax.plot(pCO2, HCO3*KILO, label='[SID] = %d mequiv./L' % sid)

pCO2_ = 41
nh  = Fh4(42*MILI, Atot, pCO2_ )
HCO3 = Kc*pCO2_/nh
ax.plot( pCO2_, HCO3*KILO, '+', color='red', label='normal operating point' )

ax.legend()
fig.tight_layout()
fig.subplots_adjust( top=0.88 )

plt.show()
```
